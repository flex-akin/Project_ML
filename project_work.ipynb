```python
def check(x):
    if x == np.nan:
        return 0 
    elif type(x) == list:
        for i in x:
            if i['name'] == 'Signal Strength ':
                return(i['value'])
            
```

```python
def check1(x):
    if x == np.nan:
        return 0 
    elif type(x) == list:
        for i in x:
            if i['name'] == 'RSRQ':
                return(i['value'])
```

```python
def check2(x):
    if x == np.nan:
        return 0 
    elif type(x) == list:
        for i in x:
            if i['name'] == 'RSRP':
                return(i['value'])
```

```python
df['Signal Strength'] = df['networkParameters'].apply(check)
df['RSRQ'] = df['networkParameters'].apply(check1)
df['RSRP'] = df['networkParameters'].apply(check2)
```
