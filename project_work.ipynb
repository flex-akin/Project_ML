```python
x = 'Lat. 7.2954142 Long. 5.1342486'

def change(y):
    if type(y) == str:
        f = y.split(" ")
        g = f[1]
        return g
    else:
        return 0
    

def changes(ya):
    if type(ya) == str:
        f = ya.split(" ")
        g = f[3]
        return g
    else:
        return 0
    
def changer(yam):
    if type(yam) == str:
        f = yam.split(" ")
        g = f[1] + "," + " " + f[3]
        return g
    else:
        return 0
```

```python
broad_band_df['Latitude'] = broad_band_df['location'].apply(change)
broad_band_df['Longitude'] = broad_band_df['location'].apply(changes)
broad_band_df['Location'] = broad_band_df['location'].apply(changer)
```
